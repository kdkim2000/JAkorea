````python
!pip install yfinance==0.2.17
import yfinance as yf
import datetime

today = str(datetime.datetime.now().date())
start_date = '2000-01-01'
end_date = today

# JPY=X for JPY/USD, KRW=X for KRW/USD, EURUSD=X for EUR/USD
currency_pairs = ['JPY=X', 'KRW=X', 'EURUSD=X']

# Get data and save to corresponding DataFrames 
for pair in currency_pairs:
    data = yf.download(pair, start=start_date, end=end_date)
    
    if pair == 'JPY=X':
        jpydf = data
        
    elif pair == 'KRW=X':
        krwdf = data
        
    elif pair == 'EURUSD=X':
        eurdf = data

````

````python
!pip install yfinance==0.2.17
import yfinance as yf
import datetime

today = str(datetime.datetime.now().date())
start_date = '2000-01-01'
end_date = today
currency_pair = 'KRW=X'

data = yf.download(currency_pair, start=start_date, end=end_date)
data
````

data 에서  "Date" 열과 "Close" 열만 따로 추출해서 data_simple 변수에 저장해주는 코드로 발전시켜줘

````python
data_simple = data[['Close']]
````

```
이제 data_simple 안에 있는 데이터를 전처리할꺼야.
전처리의 목적은 data_simple 안의 Close 열값을 한칸씩 shift시켜서 Close-1,Close-2,,,,,,,Close-30 까지 해서 총 30개의 열값을 만들어내는 코드로 발전시켜줘
만들어진 데이터는
data_preprocess 라는 변수에 넣어줘
```

````python
data_preprocess = data_simple.copy() # 원본 데이터 유지를 위해 복사본 생성

for i in range(1, 31):
    data_preprocess['Close-' + str(i)] = data_simple['Close'].shift(i)

# Drop rows with NaN values
data_preprocess.dropna(inplace=True)
````

```
이제 data_preprocess 변수에 들어있는
'Close-1','Close-2',,,,,,'Close-29','Close-30' 의 데이터를 기반으로
'Close' 값은 우리가 예측해야하는 머신러닝 코드를 만들꺼야.

Training Data는 2022.12.31 까지의 데이터로
Test Data는 2023.01.01~오늘까지의 데이터로 구분해주는 코드로 발전시켜줘
```

````python
train_data = data_preprocess[data_preprocess.index <= '2022-12-31']
test_data = data_preprocess[data_preprocess.index >= '2023-01-01']
````

```
이제 각 train_data 를 train_data_X  / train_data_y 로 나누어줘
마찬가지로 test_data 를 train_data_X  / train_data_y 로 나누어줘
```

````python
# For training data
train_data_X = train_data.drop('Close', axis=1)
train_data_y = train_data['Close']

# For testing data
test_data_X = test_data.drop('Close', axis=1)
test_data_y = test_data['Close']
````

```
이제 Randomforest Regressor를 활용해서 train_data_X(feature) 와 train_data_y를 목표변수(target)로 하여 머신러닝 모델을 학습시켜줘
```

````python
from sklearn.ensemble import RandomForestRegressor

# Create and train the model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(train_data_X, train_data_y)
````

```
자 이제 만들어진 AI모델을 기반으로 test_data_X를 입력변수(feature) 로 넣은다음 predict_y 변수에 예측치를 저장해줘
```

````python
predict_y = model.predict(test_data_X)
````

마지막으로, predict_y 와 test_data_y 의 값을 시각화하여 두 데이터의 차이를 보여줘
이때, X축은 시간의 흐름으로
y축은 predict_y와 test_data_y의 수치로 보여줘

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

# Plot the actual values
plt.plot(test_data.index, test_data_y, label='Actual', color='blue')

# Plot the predicted values
plt.plot(test_data.index, predict_y, label='Predicted', color='red')

plt.title('Actual vs Predicted')
plt.xlabel('Time')
plt.ylabel('Close Value')
plt.legend()

plt.show()
````
